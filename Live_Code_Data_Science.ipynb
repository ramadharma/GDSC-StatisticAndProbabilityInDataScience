```python
data = {
    'name': ['Karl','Boris','Theoderic','Helmut','Wolfram','Marius','Emanuelle','Aldebrand','Valmir','Alberich'],
    'jenis kelamin': ['Laki-Laki','Laki-Laki','Laki-Laki','Laki-Laki','Laki-Laki','Laki-Laki','Perempuan','Laki-Laki','Laki-Laki','Laki-Laki'],
    'usia': [32,57,50,46,34,40,27,63,62,20],
    'bb': [60.5,50.9,49.5,50,40.7,46.7,45.5,49.7,54.3,32.0],
    'pendidikan': ['Doktor','Magister','Magister','Magister','Magister','Magister','Doktor','Magister','Magister','Sarjana']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
df
```

```python
median_bb = df['bb'].median()
median_bb
```

```python
import matplotlib.pyplot as plt
from collections import Counter
```

```python
pendidikan_counts = Counter(data['pendidikan'])

fig, ax = plt.subplots()
ax.bar(pendidikan_counts.keys(), pendidikan_counts.values(), color = "orange")
ax.set_xlabel('Gelar Terakhir')
ax.set_ylabel('Jumlah Orang')
ax.set_title('Grafik Pendidikan Terakhir')
plt.show()
```

```python
import seaborn as sns
from scipy.stats import skew

# Pilih hanya kolom numerik
numerik = df.select_dtypes(include=["int64", "float64"]).columns

# Membuat plot
plt.figure(figsize=(12, 6))

for i, col in enumerate(numerik, 1):
    plt.subplot(1, len(numerik), i)  # Membuat subplot dalam satu baris
    sns.histplot(df[col], kde=True, bins=20)
    plt.title(f"{col}\nSkew: {skew(df[col]):.2f}")

plt.tight_layout()
plt.show()
```
